```python
#load the data
boston = load_boston()
```

```python
features = pd.DataFrame(boston.data,columns=boston.feature_names)
target = pd.DataFrame(boston.target, columns=["Target"])
df = pd.concat([features,target],axis=1)
```

```python
x = df['RM']
y = df['Target']

x = np.array(x/x.mean())
y = np.array(y/y.mean())

n=int(0.8*len(x))
print(n)
```

```python
x_train = x[:n]
y_train = y[:n]

x_test = x[n:]
y_test = y[n:]
```

```python
# hypothesis
def h(a,x,b):
    return a*x+b

#error function
def error(a,x,b,t):
    e=0
    m=len(x)
    for i in range(m):
        e+=np.power((h(a,x[i],b)-t[i]),2)
    return (1/(2*m)*e)

#step gradient function
def step_gradient (a,x,b,t, learning_rate):
    grad_a=0
    grad_b=0
    m=len(x)
    for i in range(m):
        grad_a += 2/m * (h(a,x[i],b)-t[i])*x[i]
        grad_b += 2/m * (h(a,x[i],b)-t[i])
    a = a -(grad_a * learning_rate)
    b = b -(grad_b * learning_rate)
    return (a,b)
    
#
def perform_descent(init_a, init_b,x,t,learning_rate,iterations):
    a= init_a
    b= init_b
    for i in range(iterations):
        if(i%1000==0):
            e = error(a,x,b,t)
            print("Error {}  a:{}, b:{}".format(e,a,b))
        a,b=step_gradient(a,x,b,t,learning_rate)
    return (a,b)
```

```python
init_a = 2.6243445310252693
init_b = -1.5715850050924138

learning_rate=.001
iterations=10000

final_a,final_b=perform_descent(init_a,init_b,x_train,y_train,learning_rate,iterations)
```

```python
# hypothesis
def h(a,x1,b,x2,c):
    return a*x1+b*x2+c

#error function
def error(a,x1,b,x2,c,t):
    e=0
    m=len(x1)
    for i in range(m):
        e+=np.power((h(a,x1[i],b,x2[i],c)-t[i]),2)
    return (1/(2*m)*e)

#step gradient function
def step_gradient (a,x1,b,x2,c,t, learning_rate):
    grad_a=0
    grad_b=0
    grad_c=0
    m=len(x1)
    for i in range(m):
        grad_a += 2/m * (h(a,x1[i],b,x2[i],c)-t[i])*x1[i]
        grad_b += 2/m * (h(a,x1[i],b,x2[i],c)-t[i])*x2[i]
        grad_c += 2/m * (h(a,x1[i],b,x2[i],c)-t[i])
    a = a -(grad_a * learning_rate)
    b = b -(grad_b * learning_rate)
    c = c -(grad_c * learning_rate)
    return (a,b,c)
    
#
def perform_descent(init_a, init_b,init_c,x1,x2,t,learning_rate,iterations):
    a= init_a
    b= init_b
    c= init_c
    for i in range(iterations):
        if(i%1000==0):
            e = error(a,x1,b,x2,c,t)
            print("Error {}  a:{}, b:{}, c:{}".format(e,a,b,c))
        a,b,c=step_gradient(a,x1,b,x2,c,t,learning_rate)
    return (a,b,c)
```
